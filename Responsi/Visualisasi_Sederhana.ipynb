Skala-skala Data
1. Nominal
Nominal merupakan skala pengukuran yang paling sederhana. Ini adalah skala yang digunakan untuk memberikan label pada variabel yang tidak memiliki nilai kuantitatif. Seperti agama, ras, preferensi politik, golongan darah. Ingat bahwa tidak ada **urutan** yang alami. Ukuran keterpusatan yang bisa digunakan adalah **modus**
2. Ordinal
Skala untuk data yang memiliki urutan yang alami tapi tidak dapat dibuat perbandingan kuantitatif. Contoh: pangkat, status sosial, jenjang pendidikan. Ingat bahwa: ada **urutan** alami, bisa digunakan ukuran keterpusatan yaitu modus dan median
3. Interval
Skala yang memiliki **urutan** alami dan perbedaan yang dapat **dikuantifikasi** tapi tidak ada **nol mutlak**. Contohnya: temperatur, rating kredit, IQ. Bisa digunakan ukuran keterpusatan yaitu mean, modus,median
4. Ratio
Skala yang memiliki **urutan** yang alami, perbedaan yang dapat **dikuantifikasi**, dan memiliki **nol mutlak**. Contohnya: Usia, tinggi badan, gaji. Bisa digunakan ukuran keterpusatan yaitu mean, modus,median

##The Functions

```python
# Create a figure with two subplots side by side

def plots_1d(df: pd.DataFrame, x_cat: str=None, x_num: str=None, rot: int = 0):
    '''
    df for the data frame
    x_cat for the column name type categorical
    x_num for the column name type numeric
    rot for rotation of the labels
    '''
    if x_cat and x_num:
        print('Choose just 1 column name')

    elif x_cat:
        fig, axes = plt.subplots(1, 3, figsize=(16, 5))
        sns.set(style="whitegrid")

        # Bar count plot
        ax1 = axes[0]
        sns.countplot(data=df, x=x_cat, ax=ax1)
        ax1.set_xticklabels(ax1.get_xticklabels(), rotation=rot)
        ax1.set_title("Bar Count Plot")

        # Pie chart
        ax2 = axes[1]
        category_counts = df[x_cat].value_counts()
        ax2.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', startangle=90)
        ax2.axis('equal')  # Equal aspect ratio for a circular pie chart
        ax2.set_title("Pie Chart")

        # Donut chart
        ax3 = axes[2]
        category_counts = df[x_cat].value_counts()
        wedges, texts, autotexts = ax3.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', startangle=90, pctdistance=0.85)
        ax3.axis('equal')  # Equal aspect ratio for a circular donut chart
        ax3.set_title("Donut Chart")

        # Add a white circle to create the donut effect
        center_circle = plt.Circle((0,0),0.70,fc='white')
        ax3.add_artist(center_circle)

        # Adjust spacing between subplots
        plt.tight_layout()
        plt.show()


    elif x_num:
        fig, axes = plt.subplots(1, 2, figsize=(12, 5))
        sns.set(style="whitegrid")

        # Histogram for numerical data
        ax1 = axes[0]
        sns.histplot(df[x_num], ax=ax1, kde=True, color='red')
        ax1.set_xticks(ax1.get_xticks())  # Explicitly set x-axis tick positions
        ax1.set_xticklabels(ax1.get_xticklabels(), rotation=rot)
        ax1.set_title("Histogram")

        # Box plot for numerical data
        ax2 = axes[1]
        sns.boxplot(data=df, y=x_num, ax=ax2, color='yellow')
        ax2.set_xticks(ax2.get_xticks())  # Explicitly set x-axis tick positions
        ax2.set_xticklabels(ax2.get_xticklabels(), rotation=rot)
        ax2.set_title("Box Plot")

        # Adjust spacing between subplots
        plt.tight_layout()
        plt.show()

    else:
      print("uh oh...")
```

```python
# Create a figure with two subplots side by side

def plots_2d(df: pd.DataFrame, x_num: str = None, x_cat: str = None, y: str = None, hue: str = None, rot: int = 0):
    '''
    df for the data frame
    x_cat for the column name type categorical x-axis
    x_num for the column name type numeric x-axis
    y for categorical y-axis
    hue is for additional variability in the visualization with column name type categorical
    rot for rotation of the labels
    '''

    sns.set(style="whitegrid")

    if x_cat and x_num and y:
        print('Choose just 1 column name')
        return

    if x_num and y:
        # Scatter plot for numerical x
        plt.figure(figsize=(12, 5))
        sns.scatterplot(data=df, x=x_num, y=y, hue=hue, color='red')
        plt.title("Scatter Plot")
        plt.show()

    elif x_cat and y:
        # Create a figure with two subplots side by side
        fig, axes = plt.subplots(1, 3, figsize=(16, 5))
        sns.set(style="whitegrid")

        # First Plot (Bar Plot)
        ax1 = axes[0]
        sns.barplot(data=df, x=x_cat, y=y, hue=hue, ax=ax1)
        ax1.set_xticks(ax1.get_xticks())  # Explicitly set x-axis tick positions
        ax1.set_xticklabels(ax1.get_xticklabels(), rotation=rot)
        ax1.set_title("Bar Plot")

        # Second Plot (Box Plot)
        ax2 = axes[1]
        sns.boxplot(data=df, x=x_cat, y=y, hue=hue, ax=ax2)
        ax2.set_xticks(ax2.get_xticks())  # Explicitly set x-axis tick positions
        ax2.set_xticklabels(ax2.get_xticklabels(), rotation=rot)
        ax2.set_title("Box Plot")

        # Third Plot (Strip Plot)
        ax3 = axes[2]
        sns.stripplot(data=df, x=x_cat, y=y, hue=hue, ax=ax3, color='magenta')
        ax3.set_xticks(ax3.get_xticks())  # Explicitly set x-axis tick positions
        ax3.set_xticklabels(ax3.get_xticklabels(), rotation=rot)
        ax3.set_title("Strip Plot")

        plt.tight_layout()
        plt.show()

    else:
        print("Have you added one of x_num or x_cat and y? if so, uh oh...")
```
